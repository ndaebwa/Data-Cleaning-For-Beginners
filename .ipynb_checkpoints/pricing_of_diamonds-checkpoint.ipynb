### importing packages and libraries

```python
# importing packages and libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### Reading the data from the dataset

```python
# reading the data 
data = pd.read_csv('pricingOfDiamonds.csv') 
data
```

```python
data.isnull().sum()
```

From the above output, we can see that Perception Change and the some values of P are missing

- The reason as to why we are checking for the types is toactually tell the best imputation technique to use.

```python
# since the values of P and PC are categorical, we are going to use mean to perform the imputations
data['P'].fillna(value = data['P'].mode()[0], inplace=True)
data['PC'].fillna(value = data['PC'].mode()[0], inplace=True)
```

```python
# cross checking to see if we doonot have any missing value left
data.isnull().sum()
```

- Removing unwanted columns for example ID is also important in data cleaning

```python
# Droping ID column as it is not required for the analysis
data.drop('ID', axis=1, inplace=True)
```

```python
#separatig the categorical and numerical variables
categoriacal_data = data.select_dtypes(include='object')
numerical_data = data.select_dtypes(exclude='object')
```

## Check for outliers for continious features

```python
#Plotting the boxplot

def boxplot(data):
    for i in data.columns:
        plt.boxplot(data[i])
        plt.title(i)
        plt.show()
        
```

```python
boxplot(numerical_data)
```

## Removing the outliers

```python


# removing the outliers from the numerical data

def remove_outliers(data):
    for i in data.columns:
        q1 = data[i].quantile(0.25)
        q3 = data[i].quantile(0.75)
        iqr = q3-q1
        ub = q3 + 1.5*iqr
        lb = q1 - 1.5*iqr
        data[i] = np.where(data[i]>ub, ub, data[i])
        data[i] = np.where(data[i]<lb, lb, data[i])

remove_outliers(numerical_data)
```

# Checking if all outliers are out

```python
# plotting the boxplot again to check if the outliers are removed or not
boxplot(numerical_data)
```

# using anova to show the relationship between the categorical and continious variables


```python
def anova(data,continious_variable,categorical_variable):
    SelectedVariables = []
    print("----ANOVA RESULTS----\n")
    for i in categorical_variable:
        CategoryGroupList = data.groupby(i)[continious_variable].apply(list)
        Anova_results = f_oneway(*CategoryGroupList)
        # if the  anova p-value is greaterv than 0.05, we reject the null hypothesis 
        if Anova_results[1] < 0.05:
            print("The variable {} is correlated with {} with a p-value of {}".format(i,continious_variable,Anova_results[1]))
            SelectedVariables.append(i)
        else:
            print("The variable {} is not correlated".format(i))
    return SelectedVariables
    
```

```python
anova(data,"price",['cut','colour','clarity','P','PC'])
```

```python
anova(data,"carat",['cut','colour','clarity','P','PC'])
```

```python
anova(data,"depth",['cut','colour','clarity','P','PC'])
```

```python
anova(data,"x",['cut','colour','clarity','P','PC'])
```

```python
anova(data,"y",['cut','colour','clarity','P','PC'])
```

```python
anova(data,"depth",['cut','colour','clarity','P','PC'])
```

### From the above tests, it is observed that P and PC are not correleted to any continious variable.

# Using comparative bar to show the relationship between the categorical and categorical variables.


```python
# comparative bar graph to show associations between categorical variables
def bar_graph(data, categorical_variable):
    for i in categorical_variable:
        sns.countplot(x=i, data=data)
        plt.show()
        
```

```python
bar_graph(data, ['cut','colour','clarity','P','PC'])
```

```python
# Alternativly using matplotlib to plot the bar graph
def bar_graph1(data, categorical_variable):
    for i in categorical_variable:
        data[i].value_counts().plot(kind='bar')
        plt.show()
        
```

```python
bar_graph1(data, ['cut','colour','clarity','P','PC'])
```
